## 2. Variabelen en Datatypes

In Python kun je variabelen gebruiken om waarden op te slaan:

```python
# Strings (tekst)
naam = "Jan"
print(f"Mijn naam is {naam}")

# Getallen
leeftijd = 16
print(f"Ik ben {leeftijd} jaar oud")

# Kommagetallen
lengte = 1.75
print(f"Mijn lengte is {lengte} meter")
```

## 3. Rekenen met Python

Python kan als rekenmachine gebruikt worden:

```python
# Optellen
print(5 + 3)

# Aftrekken
print(10 - 4)

# Vermenigvuldigen
print(6 * 7)

# Delen
print(20 / 4)

# Machtsverheffen
print(2 ** 8)  # 2 tot de macht 8
```

## 4. Lijsten

Lijsten worden gebruikt om meerdere waarden op te slaan:

```python
favoriete_vakken = ["Wiskunde", "Informatica", "Natuurkunde"]

print("Mijn favoriete vakken:")
for vak in favoriete_vakken:
    print(f"  - {vak}")
```

## 5. Functies

Functies helpen je om code te organiseren en herbruikbaar te maken:

```python
def begroet(naam):
    return f"Hallo, {naam}! Welkom bij Python!"

# Gebruik de functie
print(begroet("Student"))
print(begroet("Docent"))
```
